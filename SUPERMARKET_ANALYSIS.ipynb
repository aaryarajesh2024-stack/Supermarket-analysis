#2. Data Cleaning and Preparation

```python
#to handle and fill in the missing values
#here there are no null values in any of the columns, so there is no need to fill in the missing values
```

```python
#to convert the date column into datetime type

df["Date"]=pd.to_datetime(df["Date"])

#now to check if the datatype has been converted

df["Date"].dtype

#the datatype has been converted
```

```python
#to create a totalsales column
#here there already exists a sales column which is the totalsales column as the sales column is quantity*unit price+tax 5%
#we can create a totalsales column from sales column

df['TotalSales']=df['Sales']

#to check if the column has been created
df.head()

#both sales and totalsales columns have the same values
```

#3. Exploratory Data Analysis

```python
# to find the top 3 product lines contributing the most to totalsales

product_sales_top3=df.groupby('Product line')['TotalSales'].sum().sort_values(ascending=False).head(3)
print("The product lines contributing the most to the total sales are:",product_sales_top3)
```

```python
#to determine which branch has the highest average sales

branch_average_top=df.groupby("Branch")['TotalSales'].mean().sort_values(ascending=False).head(1)
print("The branch that has the highest average sales is:",branch_average_top)
```

```python
# to find the distribution of customer types and identify which group spends more on average.

#first to find the distribution of customer types

df["Customer type"].unique()

#there are two categories of customer type- member and normal
```

```python
#to find which customer type spends more on average

custype_avg=df.groupby("Customer type")['TotalSales'].mean().sort_values(ascending=False)
print("The customer type average sales analysis is:",custype_avg)
```

```python
#gender wise average sales analysis

gen_avgsales=df.groupby("Gender")["TotalSales"].mean().sort_values(ascending=False)
print("The gender-wise average sale analysis is:",gen_avgsales)
```

#4. Visualizations

```python
#to plot a bar chart showing total sales by product line.

product_sales=df.groupby('Product line')['TotalSales'].sum()

product_sales.plot(kind='bar')
plt.title("Total sales by each product line")
plt.xlabel("Product line")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Create a box plot to compare sales distribution across branches.

sns.boxplot(x="Branch",y="TotalSales",data=df)
plt.title("Sales distribution across branches")
plt.xlabel("Branch")
plt.ylabel("Total Sales")
plt.show()
```

```python
# to draw a histogram for the TotalSales column.

sns.histplot(df["TotalSales"])
plt.show()
```

```python
# to use Seaborn pairplot for numerical attributes to identify relationships.

sns.pairplot(df.select_dtypes(include=["float64","int64"]))
```

```python
#to plot a correlation heatmap for all numerical features.

corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
plt.title("Correlation heatmap")
plt.show()
```

```python
#to find which product line generates consistent sales regardless of branch
#using standard deviation

df.groupby('Product line')['TotalSales'].std().sort_values()
```

ANALYSIS

The product line 'Food and Beverages' contributes the highest to the overall revenue, followed by 'Sports and Travel' and 'Electronic accessories'.

The branch that has the highest average sales is 'Giza'

Members contribute more to the overall revenue than normal customers.

Which product line generates consistent sales regardless of branch?

The consistency of sales across branches can be evaluated by measuring the variation in sales for each product line. A product line with low variation(standard deviation) indicates that it performs similarly well across all branches.
The product line 'Health and Beauty' has the most consistent sales across branches because it has the lowest standard deviation. Based on the analysis, Health and Beauty product line shows the least fluctuation, indicating stable performance across all branches.

This suggests that customer demand for this product line is stable and not significantly influenced by the branch. Such consistency makes it a reliable source of revenue for the business.

Does gender have a significant influence on average sales value?

To determine whether the gender influences purchasing behaviour, the average sales value for each gender was compared. The analysis shows that the average sales value for both genders is relatively similar. The average sales value for females is slightly higher than that of males. The average sales value for females is 340.931414 whereas for males it is 299.055738. This indicates that gender does not have a strong or significant impact on the amount customers spend. Both groups contribute almost equally to overall sales. This analysis suggests that the business can focus on broader customer preferences and product demand rather than targeting a specific gender group.


Suggest one insight that could support a business decision.

First key insight from the analysis is that member customers tend to spend more on average compared to normal customers.  This suggests that the membership program is effective in encouraging higher spending behaviour . Customers who are a part of the membership program may feel more engaged or motivated due to benefits such as discounts,rewards or loyalty points. Based on this insight, the business can focus on increasing the number of member customers by promoting the advantages of membership. Strategies such as special offers,reward programs or exclusive deals for members could further enhance customer retention and increase overall revenue

Second key insight from the analysis is that certain product lines contribute significantly more to total sales compared to others. This indicates strong customer demand for these categories. The business should prioritize these high-performing product lines by ensuring adequate stock availability, improving product visibility in stores and running targeted promotions. And also, to improve the quality of the low performing product lines by reviewing customer feedback. This can help maximize revenue and improve overall sales performance.

Third key insight is that certain product lines perform well in one branch but not in others. It indicates that customer preferences vary by location. The business can adopt a more localized strategy by tailoring product offerings based on branch specific demand. This approach can improve customer satisfaction and increase sales by aligning inventory with local preferences.

Fourth key insight is that the analysis of the TotalSales histogram shows that most transactions fall within the lower sales range(under 300) and only a few transactions have very high values(above 600). This indicates that the majority of customers make small to moderate purchases,while high-value transactions are relatively rare. This pattern suggests an oppurtunity for the business to increase revenue by encouraging customers to make larger purchases through bundle offers or discounts.
